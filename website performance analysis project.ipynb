```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data-export (1).csv")
```

```python
df
```

```python
df.columns = df.iloc[0]
df = df.drop(index = 0).reset_index(drop = True)
df.columns = ["channel", "date & hour", "Users",	"Sessions",	"Engaged sessions",	"Average engagement time per session",	"Engaged sessions per user",	"Events per session",	"Engagement rate",	"Event count"]
```

```python
df
```

```python
df.info()
```

```python
df["date & hour"] = pd.to_datetime(df["date & hour"], format="%Y%m%d%H", errors="coerce")
```

```python
df.head()
```

```python
numeric_cols = df.columns.drop(["channel", "date & hour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
df["Hour"] = df["date & hour"].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## what patterns or trends can you observe in website sessions and users over time

```python
# sessions and over time
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("date & hour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users over time")
plt.ylabel("Count")
plt.show()
```

# total users by channel

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="channel", y="Users", estimator=np.sum, palette="viridis")
plt.title("Total users of channel")
plt.xticks(rotation=45)
plt.show()
```

# Average engagement time by channel

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="channel", y="Average engagement time per session", estimator=np.mean, palette="magma")
plt.title("Avg Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

# engagement rate distribution by channel

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="channel", y="Engagement rate", palette="coolwarm")
plt.title("Engagement rate distribution by channel")
plt.xticks(rotation=45)
plt.show()
```

# engaged VS non engaged session

```python
df.head()
```

```python
session_df = df.groupby("channel")[["Sessions","Engaged sessions"]].sum().reset_index()
session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]
session_df_melted = session_df.melt(id_vars="channel", value_vars=["Engaged sessions", "Non-Engaged"])

plt.figure(figsize=(8,6))
sns.barplot(data=session_df_melted, x="channel", y="value", hue="variable")
plt.title("Engaged vs Non-Engaged sessions")
plt.xticks(rotation=45)
plt.show()
```

# Traffic by hour and channel

```python
heatmap_data = df.groupby(["Hour","channel"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu",linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffic by hour and channel")
plt.xlabel("channel")
plt.ylabel("hour of day")
plt.show()
```
